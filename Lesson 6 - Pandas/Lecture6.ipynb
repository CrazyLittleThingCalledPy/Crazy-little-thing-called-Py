#### In one cell (production code should be in a .py file)

```python
import pandas as pd
kids = pd.read_csv('Kids.csv')
kids.loc[:, 'Mean Score'] = kids.loc[: , 'History':'Physics'].mean(axis = 1)
kids.sort_values('Mean Score', ascending=False, inplace=True)
kids.to_excel('SortedKids.xlsx')
```

## Playing with Android Apps

```python
apps = pd.read_csv('PlayStore.csv')
```

```python
apps.head(2)
```

```python
apps['Category'].value_counts()
```

```python
games = apps.loc[apps.loc[:, 'Category'] == 'GAME']
```

```python
games
```

```python
len(apps)
```

```python
len(games)
```

```python
games.drop_duplicates('App', inplace=True)
```

```python
games
```

```python
games.sort_values('Rating', ascending=False)
```

```python
installs_filter = games.loc[:, 'Installs_more_than'] >= 1000000
```

```python
installs_filter
```

```python
games.loc[installs_filter].sort_values('Rating', ascending=False)
```
